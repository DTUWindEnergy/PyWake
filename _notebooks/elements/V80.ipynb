# V80 example turbine

PyWake contains a few predefined turbines, e.g. the V80 from Hornsrev1

```python
import numpy as np
import matplotlib.pyplot as plt
from py_wake.examples.data.hornsrev1 import V80

wt = V80()
print('Diameter', wt.diameter())
print('Hub height', wt.hub_height())
```

```python
ws = np.arange(3,25)
plt.xlabel('Wind speed [m/s]')
plt.ylabel('Power [kW]')
plt.plot(ws, wt.power(ws),'.-')
```

```python
plt.xlabel('Wind speed [m/s]')
plt.ylabel('Ct [-]')
plt.plot(ws, wt.ct(ws))
```

The V80 object is a [`OneTypeWindTurbines`](https://topfarm.pages.windenergy.dtu.dk/PyWake/wind_turbines/OneTypeWindTurbines.html) object which is a subclass of the [`WindTurbines`](https://topfarm.pages.windenergy.dtu.dk/PyWake/wind_turbines/WindTurbines.html) class.
